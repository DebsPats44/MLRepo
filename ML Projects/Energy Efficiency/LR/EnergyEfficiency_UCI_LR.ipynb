#1) IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as tts
```

#2) READING AND WRANGLING DATA

```python
df = pd.read_csv("Energy1.csv")
```

```python
df.head()
```

```python
df.pop('Y2')
```

#3) DEFINING FEATURES AND LABELS

```python
label = df.pop("Y1")
```

```python
feature = df.to_numpy()
```

#4) TRAINING AND TESTING DATA

```python
xtrain, xtest, ytrain, ytest = tts(feature, label, test_size=0.20, random_state=42)
```

#5) INITIALISING AND FITTING DATA IN MODEL

```python
lr = LinearRegression()
```

```python
lr.fit(xtrain, ytrain)
```

#6) PREDICTING AND SCORING MODEL

```python
ypred = lr.predict(xtest)
```

```python
test = [[x1, x2, x3, x4, x5, x6, x7, x8]]
```

```python
print("Estimated Heating Load : %.4f"%lr.predict(test)[0])
print("Estimation Accuracy : {0:.2f}%".format(float((lr.score(xtest,ytest))*100)))
```
